# Bag of Words

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Load the 'abcnews-date-text' dataset
# Assuming the dataset is a CSV file, adjust the file path accordingly
df = pd.read_csv('abcnews-date-text.csv', usecols=['publish_date', 'headline_text'], nrows=1000)

# Extract the 'headline_text' column as text data
text = df['headline_text'].tolist()

# Initialize the vectorizer
vectorizer = CountVectorizer()

# Transform the text data
x = vectorizer.fit_transform(text)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Display Feature names and BOW Matrix
print('Feature names:\n', feature_names)
print()
print('BOW Matrix:\n')
print(x.toarray())
```

# N-grams

bi-gram where N = 2

```python
import pandas as pd
from nltk import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter

# Load the dataset
df = pd.read_csv('abcnews-date-text.csv', usecols=['publish_date', 'headline_text'])

# Sample: Taking a small sample for demonstration, you can use the entire dataset
df = df.sample(n=1000, random_state=42)

# Concatenate all headlines into a single string
all_headlines = ' '.join(df['headline_text'])

# Tokenize the text into words
tokens = word_tokenize(all_headlines)

# Define the N for N-grams
N = 2  # Change this to the desired N-gram size

# Create N-grams
n_grams = ngrams(tokens, N)

# Count the occurrences of each N-gram
n_grams_counts = Counter(n_grams)

# Display the most common N-grams
print(f"Top {N}-grams:")
print(n_grams_counts.most_common(10))  # Adjust the number for more or fewer top N-grams
```

# TF-IDF

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the 'abcnews-date-text' dataset
# Assuming the dataset is a CSV file, adjust the file path accordingly
df = pd.read_csv('abcnews-date-text.csv', usecols=['publish_date', 'headline_text'], nrows=10000)

# Extract the 'headline_text' column as text data
text = df['headline_text'].tolist()

# Initialize the TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Transform the text data
x_tfidf = tfidf.fit_transform(text)

# Get feature names (words)
features_names_tfidf = tfidf.get_feature_names_out()

# Display Feature names and TF-IDF Matrix
print('Feature Names(TF-IDF):\n', features_names_tfidf)
print()
print('TF-IDF Matrix:\n', x_tfidf.toarray())
```

# One-Hot Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Load the 'abcnews-date-text' dataset
# Assuming the dataset is a CSV file, adjust the file path accordingly
df = pd.read_csv('abcnews-date-text.csv', usecols=['publish_date', 'headline_text'],nrows=10000)

# Extract the 'headline_text' column as text data
text = df['headline_text'].tolist()

# Step 1 - Tokens
tokens = [word for sent in text for word in sent.lower().split()]

# Step 2 - Vocabulary
vocabulary = list(set(tokens))  # Unique words in the text

# Initialize encoder
encoder = OneHotEncoder(categories=[vocabulary], sparse=False)

# Perform one-hot encoding
one_hot_encoded = []
for sent in text:
    sent_encoded = []
    for word in sent.lower().split():
        word_index = vocabulary.index(word)
        word_vector = np.zeros(len(vocabulary))
        word_vector[word_index] = 1
        sent_encoded.append(word_vector)
    one_hot_encoded.append(sent_encoded)

for sent in one_hot_encoded:
    print(sent)
```
